```python
# Let's recreate the visualization from Chapter 2 of Claus Wilke's book
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#reference: https://www.satelliteinternet.com/resources/states-with-the-most-bigfoot-sightings/
BigFoot = {'Washington': 676, 'Oregon':254, 'West Virginia':104, 'Idaho':93, 'Montana':52, 'Wyoming':28, 'Arkansas':108, 'Alaska':22, 'Oklahoma':104,'Ohio':302}
```

```python
#separating the keys and values so they can be inputted into the bar graph
State = list(BigFoot.keys())
Sightings = list(BigFoot.values())
img = plt.imread('bigfoot.jpg')

fig, ax = plt.subplots()
# creating the bar plot
plt.bar(State, Sightings, color = ['#fb2600', '#fbcc00','#d1fb00','#0019fb','#01faf3','#00fb7b','#fb7400','#00f6fb','#97b309','#3dbf43'],width = 0.4, zorder=1)
plt.imshow(img, extent = [0,10, 0, 700], aspect = 'auto') 
plt.ylabel("No. of Big Foot Sightings", fontsize = 6)
plt.title("Crazy People in Washington",fontsize = 20)

#why is our code ugly? Can't see the bars fully because the axes are kind of skewed, ridiculous colors and picture make it hard to see, title is judgemental, yaxis label is too small, x axis label is too big so you can't entirely see the text
```
